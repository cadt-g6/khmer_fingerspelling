# Get Selected Images


```python
import sys
sys.path.append('../..')
```

```python
dataset_path = '/Volumes/SD Card/cleaned'
destination_path = '/Volumes/SD Card/lightweight'
```

```python
import json

selected_images_dict = {}
with open('../../logs/image_selection/selected_images.json') as json_file:
    selected_images_dict = json.load(json_file)

selected_images_dict
```

```python
selected_images = []

for type in selected_images_dict.keys():
    chars = selected_images_dict[type]
    for charname in chars.keys():
        images = chars[charname]
        for char_index in images:
            variants = images[char_index]
            for variant in variants:
                path = f'{type}/{charname}/{variant}'

                if "_" in variant:
                    continue

                selected_images.append(path)

selected_images
```

```python
import shutil
import pathlib

for image in selected_images:
    source = f'{dataset_path}/{image}'
    des = f'{destination_path}/{image}'

    des_parent = des.split("/")
    des_parent.pop()
    des_parent = "/".join(des_parent)

    pathlib.Path(des_parent).mkdir(parents=True, exist_ok=True)
    shutil.copy2(source, des)
```
