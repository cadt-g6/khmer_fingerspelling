```python
import sys
sys.path.append('../..')
```

```python
import os
from lib.images_selector import ImagesSelector
```

```python
dataset_path = '/Volumes/SD Card/cleaned'
```

```python
cosonant_path = f'{dataset_path}/cosonant'
independent_vowel_path = f'{dataset_path}/independent_vowel'
stack_consonant_vowel_path = f'{dataset_path}/stack_consonant'
sub_vowel_path = f'{dataset_path}/sub_vowel'
vowel_path = f'{dataset_path}/vowel'
```

```python
def extract_number(s: str):
    return int(s.split(".")[0])


selector = ImagesSelector()
```

```python
except_classes = {'vo': 7, 'sa': 7}
cosonants_groups = selector.group_each_sample(cosonant_path, except_classes)
cleaned_cosonants_groups = selector.take_three_from_each_sample(
    cosonants_groups)
cleaned_cosonants_groups
```

```python
except_classes = {'ai': 7, 'ao': 7, 'au': 7, 'ou': 7}
independents_groups = selector.group_each_sample(
    independent_vowel_path, except_classes)
cleaned_independents_groups = selector.take_three_from_each_sample(
    independents_groups)
cleaned_independents_groups
```

```python
except_classes = {'stack_vo': 7, 'stack_yo': 7}

stack_consonant_vowels_groups = selector.group_each_sample(
    stack_consonant_vowel_path,
    except_classes
)

cleaned_stack_consonant_vowels_groups = selector.take_three_from_each_sample(
    stack_consonant_vowels_groups
)

cleaned_stack_consonant_vowels_groups
```

```python
except_classes = {
    'bantak': 6,
    'salap_pi': 7,
    'sanha_sur': 7,
    'tondokhead': 7,
    'treisapt': 7
}

sub_vowel_groups = selector.group_each_sample(
    sub_vowel_path,
    except_classes
)

cleaned_sub_vowel_groups = selector.take_three_from_each_sample(
    sub_vowel_groups
)

cleaned_sub_vowel_groups
```

```python
vowel_groups = ImagesSelector().group_each_sample(vowel_path)
cleaned_vowel_groups = ImagesSelector().take_three_from_each_sample(vowel_groups)
cleaned_vowel_groups
```

```python
dicts = {
    'cosonant': cleaned_cosonants_groups,
    'independent_vowel': cleaned_independents_groups,
    'stack_consonant': cleaned_stack_consonant_vowels_groups,
    'sub_vowel': cleaned_sub_vowel_groups,
    'vowel': cleaned_vowel_groups,
}
dicts
```

```python
import json
json_data = json.dumps(dicts, indent=4)

f = open('../../logs/image_selection/selected_images.json', 'w')
f.write(json_data)
f.close()
```
