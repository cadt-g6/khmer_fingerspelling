```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
image_path = '../datasets/input/cropedandrotated/images'

df = pd.DataFrame(columns=['file_name', 'label'])
```

```python
for image in os.listdir(image_path):
    consonant = image.split("-")[1]
    char = consonant.split('.')[0]
    char = char.split('_')[0]
    df.loc[len(df.index)] = [image, char]
```

```python
from sklearn.preprocessing import LabelEncoder
labels = LabelEncoder().fit_transform(df['label'])
labels

df['encoded_label'] = labels
df.head(10)
```

```python
df1 = pd.DataFrame(columns=['filename', 'label'])
df1['filename'] = df['file_name']
df1['label'] = df['encoded_label']
df1
```

```python
df1.to_csv('data.csv',index=False)
```

```python
df2 = pd.DataFrame(columns=['label', 'class_name'])

df2['class_name'] = df['label']
df2['label'] = df['encoded_label']
df2.drop_duplicates(inplace=True)

df2
```
